```python
import numpy as np
import matplotlib.pyplot as plt
from pyclustering.cluster.kmedians import kmedians
```

```python
# generate random clustered data with 3 clusters as the users in the city
center1 = (50, 60)
center2 = (100, 20)
center3 = (140, 100)
distance = 20


x1 = np.random.normal(center1[0], distance, size=(50,))
y1 = np.random.normal(center1[1], distance, size=(50,)) 

x2 = np.random.normal(center2[0], distance, size=(30,))
y2 = np.random.normal(center2[1], distance, size=(30,)) 

x3 = np.random.normal(center3[0], 10, size=(20,))
y3 = np.random.normal(center3[1], 10, size=(20,)) 

coords = np.array([np.concatenate((x1, x2, x3)), np.concatenate((y1, y2, y3))]).T
```

```python
# display the generated users
plt.scatter(coords[:,0], coords[:,1])
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
plt.show()
```

```python
# use k-medians approximation to cluster and find the median of each cluster
k = 5
initial_cluster_centers = coords[np.random.permutation(coords.shape[0])[:k],:]
kmedians_instance = kmedians(coords, initial_cluster_centers)
kmedians_instance.process()
solutions = np.array(kmedians_instance.get_medians())
```

```python
# plot the solution
plt.scatter(coords[:,0], coords[:,1], color='lightgrey')
for point in solutions:
    plt.scatter(point[0], point[1], marker=(5, 1), color='red')
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
plt.show()
```
